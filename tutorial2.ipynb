```python
# if문
score = 85

if score >=90:
    print("학점 : A")
elif score >= 80:
    print("학점 : B")
elif score >=70:
    print("학점 : C")
else :
    print("학점 : D")
```

```python
# while 문
i = 1
result = 0

while i < 10 :
    if i%2 == 1:
        result += i
    i+=1
    # 파이썬에서는 i++이 안된다.
print(result)
```

```python
# for문
## in 이후에는 리스트, 튜플, 문자열등이 될 수 있음.
result = 0
for i in range(1,10):
    result += i
print(result)
```

```python
scores = [90, 85, 77, 65, 97]
cheating_list = {2,4}
for i in range(5):
    if (i+1) not in cheating_list and scores[i]>=80:
        print(i+1,": 합격")
        
# 또다른 표현
scores = [90, 85, 77, 65, 97]
cheating_list = {2,4}
for i in range(5):
    if (i+1) in cheating_list :
        continue
    if scores[i]>=90:
        print(i+1, "번째 학생 합격")
```

```python
# 사용자 함수
def add(a,b):
    return a+b
print(add(3,4))

#return없이도 출력 가능
def add(a,b):
    print('함수의 결과 : ', a+b)
add(4,5)
```

```python
#global 함수를 사용하면 지역변수를 만들지 않고, 함수 바깥에 선언된 변수를 바로 참조하게 된다.

a = 0

def func():
    global a
    a+=1
    
for i in range(10):
    func()

print(a)
```

```python
# 일반식
def add(a,b):
    return a + b

print(add(3,5))

#람다식
print((lambda a,b : a+b)(3,7))
```

```python
# 입력
##파이썬에서는 데이터를 입력 받을 때는 input()을 이용한다. input()의 경우 한 줄의 문자열을 입력 받도록 해준다.
## 여러개의 데이터를 입력 받을 때는 데이터가 공백으로 구분되는 경우가 많다.
## 줄바꿈이라면 int(input())을 여러번 사용하면 되는데, 공백이라면 list(map(int, input().split()))으로 표현하다.
n = int(input())

data = list(map(int,input().split()))

data.sort(reverse=True)
data
```

```python
## 만약 입력받는 데이터 수가 작다면 다음과 같은 형식으로 표현 가능
n, m, k = map(int, input().split())
```
