```python
# 2진수 체계에서는 0.9를 정확히 표현하지 못해 0.9가 되지 않는다.
a = 0.3 + 0.6
print(a)

if a == 0.9:
    print(1)
else :
    print(0)
```

```python
# 위와같은 문제점을 해결하기 위해서 round(숫자,반올림해서 나올자리까지)를 사용한다.
a = 0.6+0.3
print(round(a,1))

if a == 0.9:
    print(1)
else :
    print(0)
```

```python
#리스트 관련기타 메서드
a = [1,4,3]
print('기본 리스트 : ', a)

a.append(2)
print(a)

a.sort()
print(a)

a.sort(reverse=True)
print(a)

a.reverse()
print(a)

a.insert(2,3)
print(a)

a.remove(1)
print(a)

remove_set = [3,4]

result = [i for i in a if i not in remove_set]
print(result)
```

```python
#튜플 자료형
##튜플 자료형은 리스트와 비슷하지만 한 번 선언된 값을 변경 할 수 없고, 소괄호를 사용한다는 차이점이 있다
##튜플은 리스트에 비해 상대적으로 공간 효율적이고, 일반적으로 각 원소의 성질이 서로 다를 때 주로 사용한다.
## 다익스트라 최단 경로 알고리즘에 사용된다.
## 슬라이싱 인덱싱 가능
a = (1,2,3,4)
print(a)

print(a[0:1])
```

```python
# 사전 자료형
## 사전 자료형은 키와 값의 쌍을 데이터로 가지는 자료형이다.
## 키-쌍 값으로 구성된 데이터를 처리함에 있어서 리스트보다 훨씬 빠르게 동작한다.
## 2부와 3부의 문제들에서는 리스트보다 사전 자료형을 사용했을 때 효과적인 문제들을 다룬다
## 파이썬에서 리스트, 문자열, 튜플 등 순차적인 정보를 담는 자료형을 iterable자료형이라고 한다. in 문법은 iterable자료형에 모두 사용가능.
data = dict()
data['사과'] = 'Apple'
data['바나나'] = 'Banana'
data['수박'] = 'Watermelon'

if '수박' in data :
    print('hello watermelon')
```

```python
## 사전 자료형 관련 함수

keys_list = data.keys()
values_list = data.values()

for i in keys_list :
    print(data[i])
```

```python
# 집합 자료형
## 순서가 없기 때문에 인덱싱으로 값을 얻을 수 없고 중복을 허용 하지 않는다.
## set을 이용하거나 {}을 사용하면 표현이 됨
## 집합 자료형도 복잡도가 낮기 때문에 잘 이용하면 좋음

a= set([1,1,2,3,4,5])
print(a)

b = {1,1,1,1,3,3,3,3,4,4,4,5}
print(b)
#합집합
print(a|b)
#교집합
print(a&b)
#차집합
print(a -b)
```

```python
# 집합 자료형 관련 함수

data = set([1,2,3])
print(data)

#새로운 원소 추가

data.add(4)

#새로운 원소 여러 개 추가
data.update([5,6])

data.remove(3)

print(data)
```
